```python
def loadData(data_filter, dirName):
    data = []

    for file in os.listdir(dirName):
        data.append(loadmat(dirName+file)) 

    for person in data:
        data_filter.append(person[list(person.keys())[3]])
    
    return data_filter
```

```python
def plot(data_filter):
    labels = ['x hip','y hip','x knee', 'y hip', 'x ankle', 'y ankle', 'x foot', 'y foot']
    for i in range(len(labels)):
        plt.figure()
        for j in range(len(data_filter)):
            plt.plot(data_filter[j][:,i])
        plt.xlabel('frame')
        plt.ylabel(labels[i])
        plt.show()
```

```python
data_filter = loadData([],'Sani/sx/')
```

```python
plot(data_filter)
```
